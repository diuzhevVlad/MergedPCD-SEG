```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
feat = np.fromfile("3.npy",dtype=np.float32).reshape((-1,3))
feat.shape
```

```python
plt.scatter(feat[:,0], feat[:,1], 0.1)
```

```python
db = DBSCAN(eps=0.2, min_samples=10).fit(feat[:,:2])
labels = db.labels_                                  # −1 = noise
unique = [c for c in np.unique(labels) if c != -1] 
clusters = [np.where(labels == c)[0] for c in unique] 
```

```python
palette = plt.cm.get_cmap("tab10", len(clusters))
for k, idxs in enumerate(clusters):
    plt.scatter(feat[idxs, 0], feat[idxs, 1],s=0.2, c=np.tile(np.array(palette(k)[:3]), (len(idxs),1)))
```
